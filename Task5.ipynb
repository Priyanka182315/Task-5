```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "RTA Dataset.csv"  # Change to the correct file path
try:
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully!")
except Exception as e:
    print("Error loading dataset:", e)
    exit()

# Convert 'Time' to hour format
df["Time"] = pd.to_datetime(df["Time"], format="%H:%M:%S", errors="coerce").dt.hour

# Set Seaborn style
sns.set_theme(style="whitegrid")

# Plot 1: Accident distribution by time of day
plt.figure(figsize=(12, 6))
sns.countplot(x=df["Time"].dropna(), order=sorted(df["Time"].dropna().unique()))
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Accidents")
plt.title("Accident Distribution by Time of Day")
plt.xticks(rotation=45)
plt.show()

# Plot 2: Accident distribution by weather conditions
plt.figure(figsize=(12, 6))
sns.countplot(y=df["Weather_conditions"], order=df["Weather_conditions"].value_counts().index)
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Conditions")
plt.title("Accident Distribution by Weather Conditions")
plt.show()

# Plot 3: Accident distribution by road surface type
plt.figure(figsize=(12, 6))
sns.countplot(y=df["Road_surface_type"], order=df["Road_surface_type"].value_counts().index)
plt.xlabel("Number of Accidents")
plt.ylabel("Road Surface Type")
plt.title("Accident Distribution by Road Surface Type")
plt.show()
```
